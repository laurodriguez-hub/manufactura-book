# 2. Capacidades Predictivas del Big Data en la Industria Manufacturera

# Capacidades Predictivas del Big Data en la Industria Manufacturera

Según el numeral **2.3 — Análisis de Big Data en diferentes industrias** de la Unidad 3, la manufactura utiliza Big Data para optimizar procesos, anticipar fallas y mejorar la calidad de los productos. A partir de esa base, se identifican y amplían siete capacidades predictivas relevantes para el sector.

---

## 1. Optimización de la cadena de suministro  
(Big Data oficial del punto 2.3)

Uso de datos de inventarios, proveedores, tiempos logísticos y sensores de producción para **predecir retrasos, cuellos de botella y necesidades de abastecimiento**.  
Permite planificar la producción con mayor precisión y reducir costos operativos.

---

## 2. Mantenimiento predictivo de equipos  
(Big Data oficial del punto 2.3)

Recolección de datos de sensores (vibración, temperatura, ciclos de uso) para **prever fallas en maquinaria antes de que ocurran**.  
Ayuda a evitar paradas no programadas, aumentar la vida útil de los equipos y mejorar la continuidad operativa.

---

## 3. Mejora de la calidad del producto mediante datos de procesos  
(Big Data oficial del punto 2.3)

Monitoreo de variables críticas como presión, humedad, temperatura o velocidad de producción para **detectar variaciones y prevenir defectos**.  
Optimiza la consistencia del producto final.

# 3. Análisis Exploratorio de Datos (EDIT 2019–2020)

# Análisis Exploratorio (EDA) – EDIT Manufacturera 2019–2020  
*(Basado en el boletín oficial del DANE)*

## 1. Descripción general del Dataset  

La EDIT Manufacturera es una encuesta oficial aplicada a empresas del sector industrial colombiano. Su objetivo es medir:

- Capacidades tecnológicas  
- Actividades de innovación  
- Uso y adopción de tecnologías digitales  
- Inversión en innovación e I+D  
- Transformación productiva y tecnológica  

Aunque no contamos con una base tabular numérica, el boletín ofrece variables, categorías e indicadores suficientes para realizar un análisis exploratorio conceptual.

---

## 2. Estructura del Dataset según el boletín

Las variables se organizan en los siguientes bloques:

### **A. Características generales**
- Ubicación  
- Actividad económica  
- Tamaño de empresa  
- Empleo  
- Ventas  

### **B. Actividades de innovación**
- Innovación en productos  
- Innovación en procesos  
- Innovación organizacional  
- Innovación en mercadeo  

### **C. Inversión en Innovación**
- Gasto en innovación  
- I+D interno  
- I+D externo  
- Capacitación técnica  
- Compra de maquinaria  

### **D. Tecnologías utilizadas**
- TIC  
- Automatización  
- Sensores  
- Big Data  
- IA  
- Cloud computing  

### **E. Resultados**
- Productividad  
- Reducción de costos  
- Eficiencia energética  
- Exportaciones  

---

## 3. Tipos de Datos Identificados

| Tipo de dato | Ejemplos |
|--------------|----------|
| Categóricos | Innovó / No innovó, Adopción de tecnología |
| Ordinales | Nivel de adopción tecnológica |
| Numéricos | Porcentajes, montos, personal |
| Indicadores | % de empresas con procesos automatizados |

---

## 4. Hallazgos principales del boletín

### ● Innovación
- 35% de empresas innovaron.  
- Más innovación en procesos que en productos.

### ● Adopción tecnológica
- 76% adoptaron TIC.  
- 25% usaron computación en la nube.  
- 15% usaron analítica de datos o Big Data.  
- 8% usaron IA.

### ● Inversión
- Mayor inversión en maquinaria, capacitación e I+D.

### ● Resultados
- Las empresas innovadoras presentan mayor productividad.

---

## 5. Potencial del Dataset para Big Data

### Modelos posibles:
- Predicción de adopción tecnológica  
- Productividad futura según inversión  
- Segmentación de empresas por madurez digital  
- Riesgo de rezago tecnológico  

---

## 6. Conclusión del EDA

El PDF de la EDIT ofrece información clara para un análisis estructural del sector manufacturero.  
Aunque no existe una base tabular numérica, las variables e indicadores permiten identificar:

- Relaciones entre innovación y productividad  
- Patrones de inversión  
- Niveles de adopción tecnológica  
- Potenciales modelos predictivos  

Este análisis exploratorio demuestra la utilidad del Big Data en la comprensión del sector manufacturero colombiano.

# 5. Conclusiones Finales

En esta sección se presentan las conclusiones generales derivadas del análisis de capacidades de Big Data aplicadas a la industria manufacturera durante el periodo 2019–2020. 

- Las tecnologías de Big Data ofrecen un potencial significativo para optimizar procesos productivos.
- La integración de datos públicos permite una caracterización más completa del sector manufacturero.
- Las capacidades predictivas identificadas fortalecen la toma de decisiones basada en datos.
- La industria está avanzando hacia modelos analíticos más maduros, pero aún existe brecha tecnológica.

Estas conclusiones permiten cerrar el ciclo analítico y justifican la importancia del Big Data en este sector.
